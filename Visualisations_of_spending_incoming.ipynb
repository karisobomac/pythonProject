```python
# Imports

import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv('transactionlogs.csv')

data.head()
```

```python
# Checking for Missing Data

data.isnull().sum()
```

```python
# Cheching the data frame info

data.info()
```

```python
data.describe()
```

```python
# visualizing the historam
data.hist(bins= 30, figsize = (20,20), color = 'r')
```

```python
# visualizing the correlation matrix

data = data.corr()
data

plt.figure(figsize = (16,16))
sns.heatmap(data, annot = True)
plt.show()
```

```python
# Plotting a graph of Sender balance against Receiver balance

plt.plot(data['receiver_balance'])
plt.title('Sender balance against Receiver balance')
plt.xlabel('Receiver balance')
plt.ylabel('Sender balance')
plt.legend(['Transaction logs'])
```

```python
# Plotting a graph of Sender balance against amount

plt.plot(data['amount'])
plt.title('Sender balance against amount')
plt.xlabel('amount')
plt.ylabel('Sender balance')
plt.legend(['Transaction logs'])
```

```python
# Plotting a graph of Receiver balance against amount

plt.plot(data['receiver_age'])
plt.title('Receiver balance against amount')
plt.xlabel('amount')
plt.ylabel('Receiver balance')
plt.legend(['Transaction logs'])
```
